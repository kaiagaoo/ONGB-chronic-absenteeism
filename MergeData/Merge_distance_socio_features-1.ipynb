```python
import pandas as pd
import numpy as np
import requests
import time

ongb_df = pd.read_csv("ONGB_with_distances.csv")

# Census API key - get free one at https://api.census.gov/data/key_signup.html
CENSUS_API_KEY = "bdccfd1aa7412236176389e7625dfaf18ae74a24Y"

years = ['1718', '1819', '1920', '2021', '2122', '2223', '2324']

# ── Step 1: Build long-format address table ───────────────────────────────────
records = []
for y in years:
    zip_col = f'Zip_{y}.1'  # residential zip
    if zip_col in ongb_df.columns:
        tmp = ongb_df[['ANON_ID', zip_col]].copy()
        tmp.columns = ['ANON_ID', 'zip']
        tmp['year'] = y
        records.append(tmp)

addr_df = pd.concat(records, ignore_index=True)
addr_df = addr_df.dropna(subset=['zip'])

# Clean zip codes - remove non-numeric characters, drop bad ones
addr_df['zip'] = addr_df['zip'].astype(str).str.extract(r'(\d{4,5})')[0].str.zfill(5)
addr_df = addr_df.dropna(subset=['zip'])
addr_df = addr_df[addr_df['zip'].str.match(r'^\d{5}$')]

# Sanity check
unique_zips = addr_df['zip'].unique()
print(f"Sample zips: {addr_df['zip'].sample(10).tolist()}")
print(f"Unique zips to query: {len(unique_zips)}")

# ── Step 2: Pull census variables by zip code ─────────────────────────────────
CENSUS_VARS = {
    'B17001_001E': 'poverty_universe',
    'B17001_002E': 'below_poverty',
    'B19013_001E': 'median_household_income',
    'B23025_002E': 'labor_force',
    'B23025_005E': 'unemployed',
    'B15003_001E': 'education_universe_25plus',
    'B15003_017E': 'hs_diploma',
    'B15003_022E': 'bachelors',
    'B25064_001E': 'median_gross_rent',
    'B25077_001E': 'median_home_value',
    'B27001_001E': 'health_insurance_universe',
    'B27001_005E': 'uninsured',
    'B01003_001E': 'total_population',
}

var_string = ','.join(CENSUS_VARS.keys())
base_url = "https://api.census.gov/data/2023/acs/acs5"

zip_data = {}
for i, zip_code in enumerate(unique_zips):
    url = f"{base_url}?get={var_string}&for=zip%20code%20tabulation%20area:{zip_code}&key={CENSUS_API_KEY}"
    try:
        r = requests.get(url, timeout=10)
        if r.status_code == 200:
            data = r.json()
            if len(data) > 1:
                row = dict(zip(data[0], data[1]))
                zip_data[zip_code] = row
        if i % 50 == 0:
            print(f"Progress: {i}/{len(unique_zips)} zips queried")
        time.sleep(0.1)
    except Exception as e:
        print(f"Error on zip {zip_code}: {e}")

print(f"\nSuccessfully pulled data for {len(zip_data)} of {len(unique_zips)} zips")

# ── Step 3: Build zip-level feature dataframe ─────────────────────────────────
zip_rows = []
for zip_code, row in zip_data.items():
    def safe(key):
        val = row.get(key, None)
        return float(val) if val not in [None, '-666666666', '-999999999'] else np.nan

    pop       = safe('B01003_001E')
    pov_uni   = safe('B17001_001E')
    below_pov = safe('B17001_002E')
    lf        = safe('B23025_002E')
    unemp     = safe('B23025_005E')
    edu_uni   = safe('B15003_001E')
    hs        = safe('B15003_017E')
    bach      = safe('B15003_022E')
    hi_uni    = safe('B27001_001E')
    unins     = safe('B27001_005E')

    zip_rows.append({
        'zip': zip_code,
        'total_population': pop,
        'poverty_universe': pov_uni,
        'below_poverty': below_pov,
        'poverty_rate_pct': (below_pov / pov_uni * 100) if pov_uni else np.nan,
        'median_household_income': safe('B19013_001E'),
        'labor_force': lf,
        'unemployed': unemp,
        'unemployment_rate_pct': (unemp / lf * 100) if lf else np.nan,
        'education_universe_25plus': edu_uni,
        'high_school_plus_rate_pct': ((edu_uni - hs) / edu_uni * 100) if edu_uni else np.nan,
        'college_degree_rate_pct': (bach / edu_uni * 100) if edu_uni else np.nan,
        'median_gross_rent': safe('B25064_001E'),
        'median_home_value': safe('B25077_001E'),
        'health_insurance_universe': hi_uni,
        'uninsured_rate_pct': (unins / hi_uni * 100) if hi_uni else np.nan,
    })

zip_features_df = pd.DataFrame(zip_rows)
zip_features_df.to_csv("zip_census_features.csv", index=False)
print("Saved zip_census_features.csv")

# ── Step 4: Merge back to each student-year ───────────────────────────────────
addr_df = addr_df.merge(zip_features_df, on='zip', how='left')

# ── Step 5: Pivot back to wide format and add to ongb_df ─────────────────────
census_cols = [c for c in zip_features_df.columns if c != 'zip']

for col in census_cols:
    for y in years:
        year_data = addr_df[addr_df['year'] == y][['ANON_ID', col]].set_index('ANON_ID')
        ongb_df[f'{col}_{y}'] = ongb_df['ANON_ID'].map(year_data[col])

print(f"Final shape: {ongb_df.shape}")
ongb_df.to_csv("ONGB_with_census_features.csv", index=False)
print("Saved ONGB_with_census_features.csv")
```

```python
import pandas as pd
import numpy as np

ongb_df = pd.read_csv("ONGB_with_distances.csv", low_memory=False)
soc_df = pd.read_csv("oakland_socioeconomic_by_zip.csv")

year_map = {
    '1718': 2017, '1819': 2018, '1920': 2019,
    '2021': 2020, '2122': 2021, '2223': 2022, '2324': 2023,
}
years = ['1718', '1819', '1920', '2021', '2122', '2223', '2324']

census_cols = [
    'total_population', 'poverty_universe', 'below_poverty', 'poverty_rate_pct',
    'median_household_income', 'labor_force', 'unemployed', 'unemployment_rate_pct',
    'education_universe_25plus', 'high_school_plus_rate_pct', 'college_degree_rate_pct',
    'median_gross_rent', 'median_home_value', 'health_insurance_universe', 'uninsured_rate_pct'
]

att_cols   = [f'AttRate_{y}'     for y in years]
enr_cols   = [f'DaysEnr_{y}'    for y in years]
abs_cols   = [f'DaysAbs_{y}'    for y in years]
chron_cols = [f'ChronicAbs_{y}' for y in years]
sed_cols   = [f'SED_{y}'        for y in years]
eth_cols   = [f'Eth_{y}'        for y in years]
flu_cols   = [f'Fluency_{y}'    for y in years]

# ── Helper ────────────────────────────────────────────────────────────────────
def most_recent(df, cols):
    return df[cols].apply(
        lambda row: row.dropna().iloc[-1] if row.notna().any() else np.nan, axis=1
    )

# ── Step 1: Census features by zip+year ──────────────────────────────────────
records = []
for y in years:
    zip_col = f'Zip_{y}.1'
    if zip_col in ongb_df.columns:
        tmp = ongb_df[['ANON_ID', zip_col]].copy()
        tmp.columns = ['ANON_ID', 'zip']
        tmp['acad_year'] = y
        tmp['cal_year'] = year_map[y]
        records.append(tmp)

addr_df = pd.concat(records, ignore_index=True)
addr_df = addr_df.dropna(subset=['zip'])
addr_df['zip'] = addr_df['zip'].astype(str).str.extract(r'(\d{4,5})')[0].str.zfill(5)
addr_df = addr_df.dropna(subset=['zip'])
addr_df = addr_df[addr_df['zip'].str.match(r'^\d{5}$')]
addr_df['zip'] = addr_df['zip'].astype(int)

soc_df['zip_code'] = soc_df['zip_code'].astype(int)

addr_df = addr_df.merge(
    soc_df[['year', 'zip_code'] + census_cols],
    left_on=['cal_year', 'zip'],
    right_on=['year', 'zip_code'],
    how='left'
).drop(columns=['year', 'zip_code'])

# Build all census columns at once (avoids fragmentation warning)
new_census_cols = {}
for col in census_cols:
    for y in years:
        year_data = (addr_df[addr_df['acad_year'] == y][['ANON_ID', col]]
                     .drop_duplicates('ANON_ID')
                     .set_index('ANON_ID'))
        new_census_cols[f'{col}_{y}'] = ongb_df['ANON_ID'].map(year_data[col])

ongb_df = pd.concat([ongb_df, pd.DataFrame(new_census_cols)], axis=1)
print(f"After census features: {ongb_df.shape}")

# ── Step 2: Engineered features ───────────────────────────────────────────────
engineered = {}

engineered['has_prior_data']        = ongb_df[att_cols[:-1]].notna().any(axis=1)
engineered['prior_years_enrolled']  = ongb_df[enr_cols[:-1]].notna().sum(axis=1)
engineered['prev_days_enr']         = most_recent(ongb_df, enr_cols[:-1])
engineered['prev_days_abs']         = most_recent(ongb_df, abs_cols[:-1])
engineered['prev_att_rate']         = most_recent(ongb_df, att_cols[:-1])
engineered['prev_chronic']          = most_recent(ongb_df, chron_cols[:-1])
engineered['prior_mean_att_rate']   = ongb_df[att_cols[:-1]].mean(axis=1)
engineered['prior_min_att_rate']    = ongb_df[att_cols[:-1]].min(axis=1)
engineered['prior_chronic_count']   = ongb_df[chron_cols[:-1]].sum(axis=1)

# School-level stats for current year
school_stats = ongb_df.groupby('SiteName_2324').agg(
    school_mean_att=('AttRate_2324', 'mean'),
    school_chronic_rate=('ChronicAbs_2324', 'mean')
).reset_index()
ongb_df = ongb_df.merge(school_stats, on='SiteName_2324', how='left')

# SED dummies
current_sed = most_recent(ongb_df, sed_cols)
engineered['SED_SED']     = (current_sed == 'SED').astype(int)
engineered['SED_Not SED'] = (current_sed == 'Not SED').astype(int)
engineered['SED_Unknown'] = current_sed.isna().astype(int)

# Eth and Fluency dummies
engineered['Eth_Asian']   = (most_recent(ongb_df, eth_cols) == 'Asian').astype(int)
engineered['Fluency_EO']  = (most_recent(ongb_df, flu_cols) == 'EO').astype(int)

ongb_df = pd.concat([ongb_df, pd.DataFrame(engineered)], axis=1)
print(f"After engineered features: {ongb_df.shape}")

# ── Step 3: Save ──────────────────────────────────────────────────────────────
ongb_df.to_csv("ONGB_final.csv", index=False)
print("Saved ONGB_final.csv")

# ── Summary ───────────────────────────────────────────────────────────────────
print(f"\nCensus match rates:")
for y in years:
    matched = ongb_df[f'poverty_rate_pct_{y}'].notna().sum()
    print(f"  {y}: {matched} of {len(ongb_df)} ({matched/len(ongb_df)*100:.1f}%)")
```

```python
import pandas as pd

df = pd.read_csv("ONGB_final_clean.csv", low_memory=False)

# Convert dist_km to dist_miles
dist_km_cols = [c for c in df.columns if 'dist_km' in c]
for col in dist_km_cols:
    new_col = col.replace('dist_km', 'dist_miles')
    df[new_col] = (df[col] * 0.621371).round(2)
df = df.drop(columns=dist_km_cols)

print(f"Distance columns: {[c for c in df.columns if 'dist_miles' in c]}")
print(f"Final shape: {df.shape}")

df.to_csv("ONGB_final_clean.csv", index=False)
print("Saved ONGB_final_clean.csv")
```
