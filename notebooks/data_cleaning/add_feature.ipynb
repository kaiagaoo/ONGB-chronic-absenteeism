```python
# Filter to 2024 — abs data was collected in 2025, so use the latest available year
crime_2024 = crime[crime['year'] == 2024].drop(columns=['year']).reset_index(drop=True)
socioeco_2024 = socioeco[socioeco['year'] == 2024].drop(columns=['year', 'area_name']).reset_index(drop=True)

print("Crime 2024:", crime_2024.shape)
print("Socioeco 2024:", socioeco_2024.shape)
crime_2024.head()
```

```python
# Prepare merge tables with _res / _sch suffixes on value columns; keep zip key columns intact
crime_res    = crime_2024.add_suffix('_res').rename(columns={'zip_code_res': 'ZipResidence'})
socioeco_res = socioeco_2024.add_suffix('_res').rename(columns={'zip_code_res': 'ZipResidence'})

# Merge all four tables onto abs df
df = (df_abs
      .merge(crime_res,    on='ZipResidence', how='left')
      .merge(socioeco_res, on='ZipResidence', how='left'))

print("Final df shape:", df.shape)
new_cols = [c for c in df.columns if c not in df_abs.columns]
print("New columns added:", new_cols)
```
